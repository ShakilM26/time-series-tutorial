```python
import pandas as pd
```

```python
df = pd.read_csv('company.csv')
df
```

```python
dates = pd.date_range(start='10/04/2022', end='10/25/2022', freq='B')
dates
```

```python
df.set_index(dates, inplace=True)
df
```

```python
# Find mean 
# target ---> september 10 to september 20

df['2022-10-10':'2022-10-20'].Google_profit.mean()
```

**Using Shift method we can change values back and forward in time**

```python

df['Google_sales(Shifted)'] = df['Google_sales'].shift(1)
```

```python
df['Google_sales(difference)'] = df['Google_sales'] - df['Google_sales(Shifted)']
df
```
